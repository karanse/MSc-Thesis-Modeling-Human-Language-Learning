# Load Data & Functions

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

## Function to get data into dictionary

```python
def import_data(basepath):
    results = {}
    with os.scandir(basepath) as entries:
        for entry in entries:
             if os.path.splitext(entry)[1]==".npy":
#                 part = os.path.basename(entry).split('_')
#                 name = part[0]+"_"+part[1]+"_"+part[2]
                results[entry.name] = np.load(basepath+os.path.basename(entry))
        return results
```

```python
def plot_lines_8vars_md(data, var1,var2,var3,var4,var5,var6,var7,var8,savename, ylim1,ylim2):
    results = data
    fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, sharex=False, sharey=False)
    
    titleli  = var1.split("_")
    titlesp = var5.split("_")
    
    def title1(x):
        if x =="li":
            return "listener"
        else:
            return "speaker"
    
    def title2(x):
        if x =="acc":
            return "accuracy"
        else:
            return "loss"
    
    
    #fig.suptitle("Accuracy & loss results of listener & speaker (batches:40, lr:0.0001) ", fontsize=20)
    

    #ax1 = fig.add_axes([0,0,1,1])

    ax1.plot(results[var1], label=var1, color = 'blue')
    ax1.plot(results[var2], label=var2, color = 'red')
    ax1.set_ylabel('Accuracy',fontsize=10)
    ax1.set_xlabel('Epoch',fontsize=10)
    ax1.set_ylim(ylim1,ylim2)

  
    ax1.set_title(title1(var1.split("_")[0])+" "+title2(var1.split("_")[2]))
    ax1.tick_params(labelsize=10)
    #ax1.legend(['train', 'validaiton'])
    
    #ax2 = fig.add_axes([0,0,1,1])
    ax2.plot(results[var3], label=var3, color = 'blue')
    ax2.plot(results[var4], label=var4, color = 'red')
    ax2.set_ylabel('Accuracy',fontsize=10)
    ax2.set_xlabel('Epoch',fontsize=10)
    ax2.set_ylim(ylim1,ylim2)


    ax2.set_title(title1(var3.split("_")[0])+" "+title2(var3.split("_")[2]))
    ax2.tick_params(labelsize=10)
   # ax2.legend()
    
    #ax3 = fig.add_axes([0,0,1,1])
    ax3.plot(results[var5], label=var5, color = 'blue')
    ax3.plot(results[var6], label=var6, color = 'red')
    #ax3.set_ylim(0,7)


    ax3.set_title(title1(var5.split("_")[0])+" "+title2(var5.split("_")[2]))
    ax3.tick_params(labelsize=10)
    ax3.set_ylabel('Loss',fontsize=10)
    ax3.set_xlabel('Epoch',fontsize=10)
    #ax3.legend()
    
    #ax4 = fig.add_axes([0,0,1,1])
    ax4.plot(results[var7], label=var7, color = 'blue')
    ax4.plot(results[var8], label=var8, color = 'red')


    ax4.set_title(title1(var7.split("_")[0])+" "+title2(var7.split("_")[2]))
    ax4.tick_params(labelsize=10)
    ax4.set_ylabel('Loss',fontsize=10)
    ax4.set_xlabel('Epoch',fontsize=10)
    #ax4.legend()
    
    fig.set_figheight(10)
    fig.set_figwidth(10)
    fig.legend(    # The line objects
           labels=['train', 'validaiton'],   # The labels for each line
           loc="lower center",
           fontsize =  'x-large',
            ncol = 2)

    fig.savefig('charts_outputs/{}.png'.format(savename))
```

```python
def plot_lines_8vars_tri(var1,var2,var3,var4,var5,var6,var7,var8,savename):
    from matplotlib.ticker import FormatStrFormatter

    fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2, sharex=False, sharey=False)
    

    ax1.plot(var1, label='dyadic scenario', color = 'green')
    ax1.plot(var2, label='triadic scenario', color = 'orange')
    ax1.set_ylabel('Accuracy',fontsize=10)
    ax1.set_xlabel('Epoch',fontsize=10)
#     ax1.set_ylim(0,1)
  
    ax1.set_title("listener Test Accuracy")
    ax1.tick_params(labelsize=10)

    
    #ax2 = fig.add_axes([0,0,1,1])
    ax2.plot(var3, label='dyadic scenario', color = 'green')
    ax2.plot(var4, label='triadic scenario', color = 'orange')
    ax2.set_ylabel('Accuracy',fontsize=10)
    ax2.set_xlabel('Epoch',fontsize=10)
    ax2.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
#     ax2.set_ylim(0,0.5)

    ax2.set_title("Speaker Test Accuracy")
    ax2.tick_params(labelsize=10)

    
    #ax3 = fig.add_axes([0,0,1,1])
    ax3.plot(var5, label='dyadic scenario', color = 'green')
    ax3.plot(var6, label='triadic scenario', color = 'orange')
#     ax3.set_xlim([0, 20])
    ax3.set_title('Listener Test Loss')
    ax3.tick_params(labelsize=10)
    ax3.set_ylabel('Loss',fontsize=10)
    ax3.set_xlabel('Epoch',fontsize=10)
    
    #ax4 = fig.add_axes([0,0,1,1])
    ax4.plot(var7, label='dyadic scenario', color = 'green')
    ax4.plot(var8, label='triadic scenario', color = 'orange')
    ax4.set_ylabel('Loss',fontsize=10)
    ax4.set_xlabel('Epoch',fontsize=10)

    ax4.set_title('Speaker Test Loss')
    ax4.tick_params(labelsize=10)
    
    
    fig.set_figheight(10)
    fig.set_figwidth(10)
    fig.legend(    # The line objects
           labels=['dyadic scenario', 'triadic scenario'],   # The labels for each line
           loc="lower center",
           fontsize =  'x-large',
            ncol = 2)

    

    fig.savefig('charts_outputs/{}.png'.format(savename))
```
